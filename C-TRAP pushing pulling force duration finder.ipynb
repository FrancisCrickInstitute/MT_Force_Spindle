```python
import pandas as pd
import os

# Initialize total counters
total_positive_duration = 0
total_negative_duration = 0
total_zero_duration = 0

# List to store the results for each file
results = []

# Create the output directory if it doesn't exist
output_dir = 'output'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Loop through the file names
for i in range(1, 11):
    # Generate the file name
    file_name = f'HSET_{i}.csv'
    
    # Load the data
    data = pd.read_csv(file_name)

    # Initialize counters for the current file
    positive_duration = 0
    negative_duration = 0
    zero_duration = 0

    # Loop through the data
    for j in range(1, len(data)):
        time_interval = data['time'][j] - data['time'][j-1]
        if data['force'][j] > 0.0253:
            positive_duration += time_interval
        elif data['force'][j] < 0.01:
            negative_duration += time_interval
        else:
            zero_duration += time_interval

    # Calculate total time for the current file
    total_time = positive_duration + negative_duration + zero_duration
    
    # Calculate the percentages
    positive_percentage = (positive_duration / total_time) * 100 if total_time != 0 else 0
    negative_percentage = (negative_duration / total_time) * 100 if total_time != 0 else 0
    zero_percentage = (zero_duration / total_time) * 100 if total_time != 0 else 0

    # Add the durations from the current file to the total durations
    total_positive_duration += positive_duration
    total_negative_duration += negative_duration
    total_zero_duration += zero_duration

    # Store the results for the current file
    results.append({
        'file_name': file_name,
        'positive_duration': positive_duration,
        'negative_duration': negative_duration,
        'zero_duration': zero_duration,
        'positive_percentage': positive_percentage,
        'negative_percentage': negative_percentage,
        'zero_percentage': zero_percentage
    })

    # Print the results for the current file
    print(f"File: {file_name}")
    print(f"Time spent in positive force: {positive_duration} seconds ({positive_percentage:.2f}%)")
    print(f"Time spent in negative force: {negative_duration} seconds ({negative_percentage:.2f}%)")
    print(f"Time spent in zero force: {zero_duration} seconds ({zero_percentage:.2f}%)")
    print("-" * 40)

# Calculate total durations across all files
total_time_all_files = total_positive_duration + total_negative_duration + total_zero_duration
total_positive_percentage = (total_positive_duration / total_time_all_files) * 100 if total_time_all_files != 0 else 0
total_negative_percentage = (total_negative_duration / total_time_all_files) * 100 if total_time_all_files != 0 else 0
total_zero_percentage = (total_zero_duration / total_time_all_files) * 100 if total_time_all_files != 0 else 0

# Add the total durations and percentages to the results
results.append({
    'file_name': 'TOTAL',
    'positive_duration': total_positive_duration,
    'negative_duration': total_negative_duration,
    'zero_duration': total_zero_duration,
    'positive_percentage': total_positive_percentage,
    'negative_percentage': total_negative_percentage,
    'zero_percentage': total_zero_percentage
})

# Print the total durations across all files
print("Total durations across all files:")
print(f"Time spent in positive force: {total_positive_duration} seconds ({total_positive_percentage:.2f}%)")
print(f"Time spent in negative force: {total_negative_duration} seconds ({total_negative_percentage:.2f}%)")
print(f"Time spent in zero force: {total_zero_duration} seconds ({total_zero_percentage:.2f}%)")

# Write the results to a CSV file in the output directory
results_df = pd.DataFrame(results)
output_file_path = os.path.join(output_dir, 'durations_summary_HSET.csv')
results_df.to_csv(output_file_path, index=False)

print(f"Results have been written to '{output_file_path}'")
```
